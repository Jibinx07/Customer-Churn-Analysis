```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
# Converting TotalCharges column to numeric (some values may be blank or text)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Removing rows with missing values
df = df.dropna()
```

```python
# Counting how many customers churned vs stayed
sns.countplot(x="Churn", data=df)

# Adding title to the chart
plt.title("Customer Churn Distribution")

# Show the plot
plt.show()
```

```python
# Comparing  churn across different contract types
sns.countplot(x="Contract", hue="Churn", data=df)

# Adding title
plt.title("Churn by Contract Type")

# Show chart
plt.show()
```

```python
# Checking how monthly charges affect churn
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)

# Adding title
plt.title("Churn vs Monthly Charges")

# Show chart
plt.show()
```

```python
# Analyzing how long customers stay vs churn
sns.boxplot(x="Churn", y="tenure", data=df)

# Adding title
plt.title("Churn vs Tenure")

# Show chart
plt.show()
```

```python
# Checking if payment method affects churn
sns.countplot(x="PaymentMethod", hue="Churn", data=df)

# Rotating labels for better readability
plt.xticks(rotation=45)

# Adding title
plt.title("Churn by Payment Method")

# Show chart
plt.show()
```

```python
# Selecting numeric columns for correlation
numeric_data = df[["tenure","MonthlyCharges","TotalCharges"]]

# Creating heatmap to see relationships
sns.heatmap(numeric_data.corr(), annot=True)

# Adding title
plt.title("Correlation Matrix")

# Show chart
plt.show()
```

```python
#MACHINE  LEARNING  MODEL
```

```python
# Converting Churn column to numeric (Yes=1, No=0)
df["Churn"] = df["Churn"].map({"Yes":1, "No":0})
```

```python
# Selecting important features for prediction
X = df[["tenure","MonthlyCharges","TotalCharges"]]

# Target variable (what we want to predict)
y = df["Churn"]
```

```python
# Spliting data into training and testing 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Importing Logistic Regression model
from sklearn.linear_model import LogisticRegression

# Creating model
model = LogisticRegression()

# Training model using training data
model.fit(X_train, y_train)
```

```python
# Predict churn for test data
y_pred = model.predict(X_test)
```

```python
# Importing evaluation metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Checking accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Showing confusion matrix
print(confusion_matrix(y_test, y_pred))

# Showing detailed classification report
print(classification_report(y_test, y_pred))
```
